```python
import numpy as np
```

```python
h = 0.125
x = np.arange(start=0, stop=1.01, step=h)
f = np.array([ 0.0, 0.124670, 0.247234, 0.364902, 0.473112, 0.563209, 0.616193, 0.579699, 0.0 ])
```

```python
Ih = (f[0] + f[len(x) - 1]) / 2.0
for i in range(1, len(x) - 1):
  Ih += f[i]
Ih *= h
print("метод трапеции с h =", h)
print("I =", Ih)
```

```python
I2h = (f[0] + f[len(x) - 1]) / 2.0
# выбирается каждый второй элемент из массива значений
for i in range(2, len(x) - 1, 2):
  I2h += f[i]
I2h *= 2 * h
print("метод трапеции с h =", 2 * h)
print("I =", I2h)
```

```python
delta = (Ih - I2h) / (2**2 - 1)
print("delta: ", delta)

Irich = Ih + delta
print("с экстрополяцией Ричардсона I =", Irich)
```

```python
# Симпсон
I = 0
N = len(x)
for i in range(1, N - 1, 2):
  I += f[i - 1] + 4 * f[i] + f[i + 1]
I *= h / 3.0
print("По методу Симпсона I =",I)
```
