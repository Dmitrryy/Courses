```python
import numpy as np
```

```python
x = [    1.5,   1.6,   1.9,     2 ]
y = [ -1.345, -0.97, 0.252, 0.693 ]
# номер из семинара
# использовала для отладки программы
# x = [    0.2,   0.25,   0.27,     0.3 ]
# y = [ 0.029, -0.08, -0.122, -0.185 ]
```

```python
# Это таблица разделенных разностей
Table = np.zeros([len(x) + 1, len(x)])
# обратная интерполяция, следовательно меняем местами y и x
Table[0] = y
Table[1] = x
Table = Table.T

for col in range(2, len(Table[0])):
  for line in range(0, len(x) - col + 1):
    xi = Table[line][0]
    xi_1 = Table[line + col - 1][0]

    fi = Table[line][col - 1]
    fi_1 = Table[line + 1][col - 1]

    Table[line][col] = (fi_1 - fi) / (xi_1 - xi)

print("таблица разделенных разностей\n", Table)
```

```python
N = len(x)
b = np.array([None] * N)
# берутся верхнии коэффициенты
for i in range(0, N):
  b[i] = Table[0][i + 1]
print("b: ", b)

x0 = 0
xi = y
res = b[0]
#вычисляем полином вида P(x) = b0 + b1 * (x - x[0]) + b2 * (x - x[0]) * (x - x[1]) + ...
for i in range(1, N):
  cur = b[i]
  for k in range(0, i):
    cur = cur * (x0 - xi[k])
  res = res + cur

print(res)
```
