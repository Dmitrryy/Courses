```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.array([ 0.0,     0.25,      0.5,     0.75,      1.0,     1.25,      1.5,     1.75,      2.0 ])
f = np.array([ 1.0, 0.979915, 0.927295, 0.858001, 0.785398, 0.716844, 0.716844, 0.716844, 0.716844 ])
h = 0.25
```

# Метод трапеции

```python
N = len(x)
I_h = (f[0] + f[N - 1]) / 2.0
for i in range(1, N - 1):
  I_h += f[i]
I_h *= h
print("I_h: ", I_h)
```

```python
N = len(x)
I_2h = (f[0] + f[N - 1]) / 2.0
for i in range(2, N - 1, 2):
  I_2h += f[i]
I_2h *= 0.5
print("I_2h: ", I_2h)

delta = (I_h - I_2h) / (2**2 - 1)
print("delta: ", delta)

I_r = I_h + delta
print("I rich: ", I_r)
```

# Симпсон

```python
I = 0
N = len(x)
for i in range(1, N - 1, 2):
  I += f[i - 1] + 4 * f[i] + f[i + 1]
I *= h / 3.0
print(I)
```
